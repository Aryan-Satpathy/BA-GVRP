```python
import pandas as pd
import numpy as np 
import warnings
import os

warnings.filterwarnings("ignore")
```

```python
dataset = {}
instance = 0
path = os.getcwd() + '\Large'
for filename in os.listdir(path):
    print(filename)
    with open(os.path.join(path, filename), 'r') as f: 
        try:
            df = pd.read_csv(f,sep='\t',header=(1))
        except pd.errors.ParserError:
            print(filename)
            continue
        df = pd.DataFrame([list(df.columns)], columns=df.columns).append(df)
        try:
            df.columns = ['ID', 'Type', 'Longitude', 'Latitude']
        except ValueError:
            print(filename)
            continue
        df.dropna(inplace=True)
        df.Longitude = df.Longitude.astype(float)
        df.Latitude = df.Latitude.astype(float)
        df.drop_duplicates(subset=['Longitude', 'Latitude'], keep='first', inplace=True)
        config = df.Type.value_counts()

        # https://stackoverflow.com/questions/24617013/convert-latitude-and-longitude-to-x-and-y-grid-system-using-python
        # dx = (lon1-lon2)*40000*math.cos((lat1+lat2)*math.pi/360)/360
        # dy = (lat1-lat2)*40000/360
        df['x'] = (df.Longitude - float(df.iloc[[0]].Longitude))*40000*np.cos((df.Latitude + float(df.iloc[[0]].Latitude))*np.pi/360)/360
        df['y'] = (df.Latitude - float(df.iloc[[0]].Latitude))*40000/360
        df['Type'] = pd.Categorical(df['Type'], ["d", "c", "f"])
        df.sort_values("Type", inplace=True)
        df['distance'] = df[["x","y"]].apply(tuple, axis=1)
        dataset[instance] = {'depo': config.d, 'customers':config.c, 'afs':config.f, 'dataset':list(df['distance'])}

        instance+=1

```
